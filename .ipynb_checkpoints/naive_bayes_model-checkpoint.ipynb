```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import style
style.use("ggplot")
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
mapping = {'FALSE POSITIVE': 0, 'CONFIRMED': 1, 'CANDIDATE': 2}
df = df.replace({'koi_disposition': mapping})
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec',
       'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 
       'koi_prad', 'koi_teq', 'koi_insol','koi_model_snr', 'koi_tce_plnt_num',
       'koi_steff',  'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag']]

X = selected_features
```

```python
target = df['koi_disposition']
target_names = ['False Positive', 'Confirmed', 'Candidate']
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_features, target, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler

# Create a minmaxscalar model and fit it to the training data
X_scaler = MinMaxScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Training - Gaussian Naive Bayes Model



```python
from sklearn.naive_bayes import GaussianNB 
gnb_model = GaussianNB()
gnb_model.fit(X_train_scaled, y_train)
gnb_prediction = gnb_model.predict(X_test_scaled)
```

```python
from sklearn import metrics 
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, gnb_prediction)*100)
```

```python
print(f"First 10 Predictions:   {gnb_prediction[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, gnb_prediction)
print('Confusion Matrix\n')
print(confusion)
```

```python
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns
# Plot the confusion matrix chart
class_names = target_names
plot_confusion_matrix(gnb_model, X_test_scaled, y_test,
                                 display_labels=class_names,cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Naive Bayes Model")
plt.show()
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
import numpy as np

param_grid = {
    'priors': [None],
    'var_smoothing': [0.00000001, 0.000000001, 0.00000001]
}

# param_grid = {'var_smoothing': np.logspace(0,-9, num=100)}
  
grid = GridSearchCV(gnb_model, param_grid,    # use any cross validation technique 
                 verbose=1, 
                 scoring='accuracy')
```

```python
gnb_model.get_params().keys()
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test_scaled)
```

```python
print('Test Acc: %.3f' % grid.score(X_test_scaled, y_test))
```

```python
#Classification Report for Gaussian Naïve Bayes 
from sklearn.metrics import classification_report
print(classification_report(y_test, gnb_prediction,
                           target_names=target_names))
```

The model is 98% accurate in predicting the false positive cases, but it is not a good option in predicting the confirmed category which is 86%, and candidate category which is only 44% accurate.
So this model is not a perfect model for this dataset.
