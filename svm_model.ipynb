# Read the CSV and Perform Basic Data Cleaning

```python
exo_data = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
exo_data = exo_data.dropna(axis='columns', how='all')
# Drop the null rows
exo_data = exo_data.dropna()
exo_data.head()
```

# Select your features (columns)

```python
# df = exo_data.drop(['koi_disposition', 'koi_period_err2'], axis=1)
```

```python
# Set features. This will also be used as your x values.
selected_features = exo_data[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec',
       'koi_period', 'koi_period_err1', 'koi_time0bk', 'koi_time0bk_err1',
       'koi_time0bk_err2', 'koi_impact', 'koi_impact_err1', 'koi_impact_err2',
       'koi_duration', 'koi_duration_err1', 'koi_duration_err2', 'koi_depth',
       'koi_depth_err1', 'koi_depth_err2', 'koi_prad', 'koi_prad_err1',
       'koi_prad_err2', 'koi_teq', 'koi_insol', 'koi_insol_err1',
       'koi_insol_err2', 'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff',
       'koi_steff_err1', 'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1',
       'koi_slogg_err2', 'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'ra',
       'dec', 'koi_kepmag']]
X = selected_features
X
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = exo_data['koi_disposition']
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
# for label, original_class in zip(encoded_y, y):
#     print('Original Class: ' + str(original_class) + '    ' + 'Encoded Label: ' + str(label))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from keras.utils import to_categorical
# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScalar model and fit it to the training data
X_scaler = MinMaxScaler().fit(X_train)
```

```python
X_train.head()
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

#### svm

```python
# Support vector machine linear classifier
from sklearn import set_config
from sklearn.svm import SVC 
svmodel = SVC(kernel='linear', verbose=2)
set_config(print_changed_only=False)
print(svmodel)
```

```python
svmodel.fit(X_train_scaled, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % svmodel.score(X_test, y_test))
```

```python
predicted_class = svmodel.predict(X_test)
print(predicted_class)
```

# Train the Model



```python
print(f"Training Data Score: {svmodel.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {svmodel.score(X_test_scaled, y_test)}")
```

```python
target = exo_data['koi_disposition']
target_names = ['Candidate', 'Confirmed', 'False Positive']
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
# grid_params = { 'n_neighbors' : [3,5,11,19], 
#                'weights': ['uniform', 'distance'] }
# grid = GridSearchCV(svmodel, grid_params, verbose=1, cv=3, n_jobs=-1)

param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf']}  
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3) 

# grid_params = { 'n_neighbors' : [3,5,11]  }
# grid = GridSearchCV(KNeighborsClassifier(), grid_params, verbose=3, n_jobs=-1)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test_scaled)
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

This svm model prediction is better than both logisitc regression and knn models. This is 98% accurate in predicting the 'False positive' category, 81% accurate in 'Candidate' category and 79% accurate in predicting 'Confirmed' category
